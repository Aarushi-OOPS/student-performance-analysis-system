```python
import numpy as np

students = np.array(["Aarushi", "Riya", "Neha", "Karan", "Rahul"])

marks = np.array([
    [85, 78, 92],
    [76, 88, 80],
    [90, 91, 89],
    [45, 50, 40],
    [60, 72, 68]
])
# Subjects:
# Math
# Science
# English
```

```python
# Print all the student names
print("Student Name",students)
```

```python
# total marks of all student
total_marks = marks.sum(axis = 1) 
print("Total_marks:", total_marks)
```

```python
# Print total marks of each student
print(f"Total_marks of {students[0]} is {total_marks[0]}")
print(f"Total_marks of {students[1]} is {total_marks[1]}")
print(f"Total_marks of {students[2]} is {total_marks[2]}")
print(f"Total_marks of {students[3]} is {total_marks[3]}")
print(f"Total_marks of {students[4]} is {total_marks[4]}")
```

```python
# Using loop
for i in range(len(students)):
    print(f"{students[i]}'s marks: {total_marks[i]}")
```

```python
# Print average marks of each student
avg_marks = marks.mean(axis=1)
print(np.round(avg_marks,2))
```

```python
# Find topper of the class
Topper = total_marks.max(axis = 0)
print(Topper)
```

```python
topper_index = total_marks.argmax(axis = 0)
print(f"Topper: {students[topper_index]}")
print(f"Marks: {total_marks[topper_index]}")
```

```python
# Find student with lowest marks
lowest_index = total_marks.argmin()
print(f"Name(lowest_marks): {students[lowest_index]}")
print(f"Marks: {total_marks[lowest_index]}")
```

```python
# Print subject-wise average
# Meaning:
# average marks in Math
# average marks in Science
# average marks in English
avg_marks_sw= marks.mean(axis=0)
print(avg_marks_sw)
```

```python
subjects = np.array(["Maths","Science","English"])
for i in range(len(avg_marks_sw)):
    print(f"Average marks in {subjects[i]} is {avg_marks_sw[i]}")
```

```python
# Find highest marks in each subject
highest_marks = marks.max(axis = 0)
print(highest_marks)
```

```python
mask = avg_marks>80
print(mask)
for i in range (len(students)):
    if mask[i]== True:
        print((f"{students[i]} got above average mark 80: {np.round(avg_marks[i],2)}",))
```

```python
mask = avg_marks<40
print(mask)
for i in range (len(students)):
    if mask[i]== True:
        print((f"{students[i]} got fail: {np.round(avg_marks[i],2)}",))
```

```python
# Calculate percentage of each student
max_marks = 300
per_mark = (total_marks/max_marks)*100
for i in range (len(students)):
    print(f"{students[i]}'s Percentage: {per_mark[i]:.2f}%")
```

```python
for i in range(len(students)):
     if (per_mark[i]>= 85):
         grade = "A"
         print(f"{students[i]} got {grade} grade")
     elif (70<=per_mark[i]< 85):
         grade = "B"
         print(f"{students[i]} got {grade} grade")
     elif (50<=per_mark[i]< 70):
         grade = "C"
         print(f"{students[i]} got {grade} grade")    
     else:
         grade = "D"
         print(f"{students[i]} got {grade} grade") 
        
```

```python
#Broadcasting
grace_marks = marks+5
print(grace_marks)
```
